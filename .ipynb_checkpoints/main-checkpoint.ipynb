### `Loading Dataset`

```python
# happy - 0
# sad - 1
data = tf.keras.utils.image_dataset_from_directory(data_dir, color_mode = 'grayscale', batch_size = 8)
```

```python
data_iterator = data.as_numpy_iterator()
```

```python
batch = data_iterator.next()
```

```python
batch[1]
```

```python
fig, ax = plt.subplots(ncols = 4, figsize=(20,20))
for idx, img in enumerate(batch[0][:4]):
    ax[idx].imshow(img.astype(int))
    ax[idx].title.set_text(batch[1][idx])
```

### `Preprocessing Dataset`

```python
data = data.map(lambda x,y : (x/255.0, y))
```

```python
len(data)
```

```python
train_size = int(len(data) * .7)
val_size = int(len(data) * .2)
test_size = int(len(data) * .1)
```

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size).skip(val_size).take(test_size)
```

### `Building a CNN Model`

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = keras.Sequential([
    keras.Input(shape = (256,256,1)),
    layers.Conv2D(16, (3,3), padding='valid', activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(32, (3,3), activation='relu'),
    layers.MaxPooling2D(),
    layers.Conv2D(16, (3,3), activation='relu'),
    layers.MaxPooling2D(),
    layers.Flatten(),
    layers.Dense(256, activation = 'relu'),
    layers.Dense(1, activation = 'sigmoid')
])
```

```python
model.summary()
```

```python
model.compile('adam', loss = tf.losses.BinaryCrossentropy(), metrics = ['accuracy'])
```

```python
 logsdir = "C:\\Users\\ACER\\Desktop\\new_emotions\\logs"
```

```python
 tensorboard_callbacks = tf.keras.callbacks.TensorBoard(log_dir = logsdir)
```

```python
hist = model.fit(train, epochs = 10, validation_data = val, callbacks = [tensorboard_callbacks])
```

```python
hist.history
```

```python
plt.plot(hist.history['accuracy'], marker = 'o', label = 'Accuracy')
plt.plot(hist.history['val_accuracy'], marker = '.', label = 'Validation Accuracy')
plt.title('Accuracy')
plt.legend()
```

### `Evaluating Performance`

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
rec = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    X, y = batch
    yhat = model.predict(X)
    pre.update_state(y, yhat)
    rec.update_state(y, yhat)
    acc.update_state(y, yhat)
```

```python
pre.result()
```

```python
rec.result()
```

```python
acc.result()
```
